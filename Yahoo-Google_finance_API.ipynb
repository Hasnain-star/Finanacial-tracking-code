```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import pandas as pd
```

### Indexes of Apple, Microsoft and the S&P500

```python
ticks = ['AAPL', 'MSFT', '^GSPC']

# set data period
start_date = '2012-08-03'
end_date = '2016-12-31'

# loading the desired data
data_panel = web.DataReader(ticks, 'yahoo', start_date, end_date)
```

```python
data_panel
```

```python
data_panel.info()
```

### Working with the close and adjusted prices

```python
data_panel.head(10)
```

```python
# getting only adjusted closing prices
close = data_panel.Close
close.head(10)
```

```python
# getting only business days between 01-01-2000 and 12-31-2016
weekends = pd.date_range(start = start_date, end = end_date, freq = 'B')
weekends
```

```python
# align all the prices with new dates (5 days a week)
close = close.reindex(weekends)
close
```

```python
# show all how much NaN values
close.isnull().sum()
```

```python
# fill NaN cells by replacing with latest available price
close.fillna(method = 'ffill', inplace = True)
close.isnull().sum()
```

### MSFT time-series

```python
# select only MSFT column
msft = close.loc[:, 'MSFT']
msft.head()
```

```python
# average moving price by rolling for 20 and 100 days
short_rolling_msft = msft.rolling(20).mean()
long_rolling_msft = msft.rolling(100).mean()
```

```python
short_rolling_msft, long_rolling_msft
```

```python
# plot of rollings
fig, ax = plt.subplots(figsize = (20, 11))

ax.plot(msft.index, msft, label='MSFT')
ax.plot(short_rolling_msft.index, short_rolling_msft, label='20 days rolling')
ax.plot(long_rolling_msft.index, long_rolling_msft, label='100 days rolling')

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted closing price ($)')
ax.legend()
```
